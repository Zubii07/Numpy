### Practice with real world data

```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
## Data Structure: [restaurant_id, month1_sales, month2_sales, month3_sales, month4_sales]
import numpy as np
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 160000, 190000, 230000, 260000],
    [3, 170000, 200000, 240000, 270000],
    [4, 180000, 210000, 250000, 280000],
    [5, 190000, 220000, 260000, 290000],
  
])

print("==== Sales Analysis ====")
print("\n Sales Data Shape:", sales_data.shape)
print("\n Sales data for first 3 restaurants:\n", sales_data[:3]) # we can use [:, 1:] to get all sales data without restaurant_id
```

```python
## total sales per year
total_Sales = np.sum(sales_data[:, 1:], axis=0)
print("\n Total Sales per year:", total_Sales)
```

```python
## Minimum sales per restaurant
min_Sales = np.min(sales_data[:, 1:], axis=1)
print("\n Minimum Sales per year:", min_Sales)

## Maximum sales per year
max_Sales = np.max(sales_data[:, 1:], axis=0)
print("\n Maximum Sales per year:", max_Sales)

## Choose axis=0 => column-wise operation(top to bottom)
## Choose axis=1 => row-wise operation(left to right)

## Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("\n Average Sales per restaurant:", avg_sales)


## Cumulative sales per restaurant
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("\n Cumulative Sales per restaurant:\n", cumsum)

## Plotting the sales data
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Cumulative Sales per Restaurant")
plt.xlabel("year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
